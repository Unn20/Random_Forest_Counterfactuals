```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import time
import pickle
from subprocess import call

from sklearn import preprocessing
from sklearn.tree import export_graphviz
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score

DATASET_PATH = "../datasets/adult/"
# CHUNKS = [i*10 for i in range(1, 10]
# CHUNKS = [i*2 for i in range(1, 11)]
# CHUNKS = [i*1 for i in range(1, 21)]
CHUNKS = [i*5 for i in range(1, 11)]
CHUNKS
```

```python
adult_dataset = pd.read_csv(DATASET_PATH + "adult_modified.csv")

feature_names = [c for c in adult_dataset.columns if c != "class"]
class_names = list(adult_dataset["class"].unique())

print(f"Positive/all ratio: {sum(adult_dataset['class']==1)/len(adult_dataset)}")

adult_dataset
```

```python
for chunk in CHUNKS:
    chunked_dataset, _ = train_test_split(adult_dataset, train_size=chunk/100, stratify=adult_dataset['class'])
    chunked_dataset.to_csv( f"./adult_part_{chunk}.csv", index=False)
    print(f"Saved to: ./adult_part_{chunk}")
```
